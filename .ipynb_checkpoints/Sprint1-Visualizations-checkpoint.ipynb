```python
import matplotlib.pyplot as plt
import config.env
import seaborn
import mysql.connector
import pandas as pd
import database.connect
import globals
```

```python
seaborn.set()
```

```python
connection=database.connect.getConnection()
cursor=connection.cursor()
```

```python
tweets_per_airline={}
```

```python
for i in range(len(globals.airlineIDs.values())):
    airlineName=list(globals.airlineIDs.keys())[i]
    airlineID=list(globals.airlineIDs.values())[i]
    countQuery="SELECT count(id) from tweets WHERE user_id=%s"
    execute=cursor.execute(countQuery,[airlineID])
    result=cursor.fetchall()
    tweets_per_airline[airlineName]=result[0][0]
print(tweets_per_airline)
```

```python
tweets_received_per_airline={}
```

```python
for i in range(len(globals.airlineIDs.values())):
    airlineName=list(globals.airlineIDs.keys())[i]
    airlineID=list(globals.airlineIDs.values())[i]
    countQuery="SELECT COUNT(id) FROM `tweets` WHERE user_mentions LIKE '%{}%';".format(airlineID)
    execute=cursor.execute(countQuery)
    result=cursor.fetchall()
    tweets_received_per_airline[airlineName]=result[0][0]
print(tweets_received_per_airline)
```

```python
fig,ax = plt.subplots(figsize=(10,5))
plt.bar(tweets_received_per_airline.keys(), tweets_received_per_airline.values())
```

```python
seaborn.barplot(pd.DataFrame({'name':tweets_per_airline.keys(),'tweet_count':tweets_per_airline.values()}),x='name',y='tweet_count',color='#354651')
df_airline_tweets_count=pd.DataFrame({'name':tweets_per_airline.keys(),'tweet_count':tweets_per_airline.values()})
df_airline_tweets_received_count=pd.DataFrame({'name':tweets_received_per_airline.keys(),'tweets_received_count':tweets_received_per_airline.values()})
df_airline_tweets_count.to_csv('output/airline_tweets_count.csv')
df_airline_tweets_received_count.to_csv('output/airline_tweets_received_count.csv')
```

```python
cursor.fetchall()
dayTweetsQuery="SELECT FLOOR((timestamp-1546300800000)/86400000)  as day ,COUNT(id),timestamp from tweets GROUP BY FLOOR((timestamp-1546300800000)/86400000)"
execute=cursor.execute(dayTweetsQuery)
result=cursor.fetchall()
tweets_per_day=[]

import datetime
begining2019 = datetime.datetime.strptime('1/1/2019', "%m/%d/%Y")

for i in result:
    tweets_per_day.append([int(i[0]),i[1],begining2019+datetime.timedelta(days=int(i[0]))])
```

```python
americanDayATweetsQuery="SELECT FLOOR((timestamp-1546300800000)/86400000)  as day ,COUNT(id),timestamp from tweets WHERE user_mentions LIKE '%22536055%' GROUP BY FLOOR((timestamp-1546300800000)/86400000)"
execute=cursor.execute(americanDayATweetsQuery)
americanResult=cursor.fetchall()
american_tweets_per_day=[]

for i in americanResult:
    american_tweets_per_day.append([int(i[0]),i[1],begining2019+datetime.timedelta(days=int(i[0]))])
```

```python
df_tweets_per_day=pd.DataFrame({'date':[i[2] for i in tweets_per_day],'tweets_count':[i[1] for i in tweets_per_day]})
df_tweets_per_day.to_csv('output/tweets_per_day.csv')
fig,ax=plt.subplots(figsize=(10,5))
seaborn.lineplot(data=df_tweets_per_day,x='date',y='tweets_count',ax=ax)
df_american_tweets_per_day=pd.DataFrame({'date':[i[2] for i in american_tweets_per_day],'tweets_count':[i[1] for i in american_tweets_per_day]})
seaborn.lineplot(data=df_american_tweets_per_day,x='date',y='tweets_count',ax=ax)
```
